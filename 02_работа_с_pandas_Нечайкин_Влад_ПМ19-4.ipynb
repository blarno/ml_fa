```python
data = pd.read_csv('titanic.csv',delimiter = ',',index_col =['PassengerId']).drop(['Unnamed: 0'],axis = 1)
```

```python
from collections import Counter
for i in data.columns:
    if(len(Counter(data[i]))<=5):
        print(i)
        print(Counter(data[i]))
```

```python
data.info()
```

```python
data = data.drop(['Name','Cabin','Ticket'],axis = 1 )
```

```python
data.head()
```

```python
data['Birth year'] = 1912 - data['Age'] 
```

```python
data['Birth year']
```

```python
data.rename(columns = {'PassengerId':'Айди', 'Survived':'Выжил','Pclass':'Класс',\
                       'Sex':'Пол','Age':'Возраст','SibSp':'Кол-во Родственников на борту',\
                       'Parch':'Кол-во Детей/родителей на борту','Ticket':'Билет','Fare':'Плата',\
                      'Cabin':'Каюта','Embarked':'Сели','Birth year':'Год рождения'}, inplace = True)
```

```python
data
```

```python
data['Возраст'] = data['Возраст'].fillna(data['Возраст'].mean())
data['Плата'] = data['Плата'].fillna(data['Плата'].mean())
data['Год рождения'] = data['Год рождения'].fillna(data['Год рождения'].mean())
data['Сели'] = data['Сели'].fillna('S')
```
